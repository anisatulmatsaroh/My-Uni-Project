# Prepocessing

```python
#Import library untuk proses preprocessing
pd.set_option('display.max_colwidth', None)
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
```

```python
#Tahap Case folding
df['Tweet'] = df['Tweet'].str.lower()
df
```

```python
#Tahap Cleansing
def cleansing(tweet):
    tweet_clean = re.sub(r'(http\S+|https\S+|twitter.com\S+|pic.twitter\S+)','',tweet)
    tweet_clean = re.sub('(@[A-Za-z0-9]+)|(#[A-Za-z0-9]+)|([^A-Za-z \t])|(\w+:\/\/\S+)', '', tweet)
    return tweet_clean
```

```python
df['Tweet'] = df['Tweet'].apply(cleansing)
df
```

```python
#Tahap hapus karakter double atau lebih
#menghilangkan huruf yang berulang( tiga atau lebih karakter)
def replaceThreeOrMore(tweet):
    pattern = re.compile(r"(.)\1{1,}", re.DOTALL)
    pattern = re.compile(r"(.)\1\1{1,}", re.DOTALL)
    pattern = re.compile(r"(.)\1\1\1{1,}", re.DOTALL)
    return pattern.sub(r"\1", tweet)
```

```python
df['Tweet'] = df['Tweet'].apply(replaceThreeOrMore)
df
```

```python
#Tahap Tokenization
def tokenize(tweet):
    return word_tokenize(tweet)
```

```python
df['Final_Tweet'] = df['Tweet'].apply(tokenize)
df
```

```python
#Tahap Stemming
factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stemming(tweet):
    return [stemmer.stem(word) for word in tweet]
```

```python
df['Final_Tweet'] = df['Final_Tweet'].apply(stemming)
df
```

```python
#Tahap mengubah kata Slangword
def convertToSlangword(tweet):
    kamus_slangword = eval(open("slangword.txt").read()) # Membuka dictionary slangword
    pattern = re.compile(r'\b( ' + '|'.join (kamus_slangword.keys())+r')\b') # Search pola kata (contoh kpn -> kapan)
    content = []
    for word in tweet:
        filteredSlang = pattern.sub(lambda x: kamus_slangword[x.group()],word) # Replace slangword berdasarkan pola review yg telah ditentukan
        content.append(filteredSlang.lower())
    tweet = content
    return tweet
```

```python
df['Final_Tweet'] = df['Final_Tweet'].apply(convertToSlangword)
df
```

```python
#Tahap Stopword removal
factory = StopWordRemoverFactory()
more_stopword = open('stopword.txt', 'r').read().split()
stopword = factory.create_stop_word_remover()
stopwords = factory.get_stop_words()+more_stopword
stopwords.remove('tidak')
#print(stopwords)
```

```python
def stop_removal(tweet):
    return [word for word in tweet if word not in stopwords]
```

```python
df['Final_Tweet'] = df['Final_Tweet'].apply(stop_removal)
df
```

# Support Vector Machine - PSO

```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import KFold
from sklearn.model_selection import train_test_split
from sklearn import metrics
import pyswarms as ps #library untuk seleksi fitur
from pyswarms.utils.plotters import (plot_cost_history, plot_contour, plot_surface)
from pyswarms.discrete.binary import BinaryPSO
from pyswarms.utils.plotters.formatters import Mesher

from sklearn.metrics import confusion_matrix
from sklearn.model_selection import RandomizedSearchCV
from scipy.sparse import csr_matrix as matrixTransform
import time as time
```

```python
#membuat definisi untuk klasifikasi SVM
classifier = SVC(C=3, gamma='scale', kernel='rbf', probability=True)
X = tfidf.values
y = label
print(X.shape)
```

```python
#Tentukan fungsi tujuan
def f_per_particle(m, alpha):
    
    total_features = nfitur
    #Dapatkan subset fitur dari biner mask
    if np.count_nonzero(m) == 0:
        X_subset = X
    else:
        X_subset = X[:,m==1]
        
    X_train, X_test, y_train, y_test = train_test_split(X_subset, y, test_size=0.2, random_state=42, shuffle=True)
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    P = (y_pred == y_test).mean()
    #Hitung untuk fungsi tujuan
    j = (alpha * (1.0 - P)#nilai fitness
        + (1.0 - alpha) * (1 - (X_subset.shape[1] / total_features)))
    return j
```

```python
def f(x, alpha=0.99):
    n_particles = x.shape[0]
    j = [f_per_particle(x[i], alpha) for i in range(n_particles)]
    return np.array(j)
```

```python
#mulai menghitung waktu pemrosesan
start=time.time()

iters = 10

#inisialisasi swarm
options = {'c1':2.1, 'c2':1.9, 'w':0.5, 'k':10, 'p':2}

#memanggil PSO
nsampel, nfitur = X.shape
dimensions = nfitur # dimensi harus sama dengan jumlah fitur
optimizer = BinaryPSO(n_particles=10, dimensions=dimensions, options=options)

#proses optimasi
cost, pos = optimizer.optimize(f, iters=iters)

end = time.time()
waktu= end-start

#waktu yang dihabiskan
print('waktu yang dihabiskan:', (waktu), "detik")
```

```python
#Menampilkan feature terpilih
print("Jumlah fitur awal : ", nfitur)
print("Jumlah fitur terpilih : ", pos.sum())
```

```python
cm = []
acc = []
recall = []
presisi = []
best_acc = 0
best_fold = 0

#mendapatkan fitur terpilih dari posisi akhir
X_selected_features = X[:,pos==1]

# Split data menjadi train dan test dengan rasio 80:20
X_train, X_test, y_train, y_test = train_test_split(X_selected_features, y, test_size=0.2, random_state=42)

#kfold
kf = KFold(n_splits=10, random_state=42, shuffle=True)
for i, (train_index, test_index) in enumerate (kf.split(X_train)):
    X_train_fold, X_test_fold, y_train_fold, y_test_fold = X_train[train_index], X_train[test_index], y_train[train_index], y_train[test_index]

    Test_X_final = matrixTransform(X_test_fold)
    Train_X_final = matrixTransform(X_train_fold)
    
    #membuat klasifikasi svm kedua
    classifier = SVC(C=3, gamma='scale', kernel='rbf', probability=True)
    #Lakukan klasifikasi dan simpan kinerja di P
    classifier.fit(Train_X_final, y_train_fold)
    predictions_svm = classifier.predict(Test_X_final)
    
    #Confusion Matrix
    #nilai 0 = negatif, nilai 1 = positif
    cf_matrix = confusion_matrix(y_test_fold, predictions_svm)
    TN = cf_matrix[0][0] #prediksi salah, aktual salah
    FN = cf_matrix[1][0] #prediksi benar, aktual salah
    TP = cf_matrix[1][1] #prediksi benar, aktual benar
    FP = cf_matrix[0][1] #prediksi salah, aktual benar
    try :
        precision_score = (TP/(TP+FP))*100
    except :
        precision_score = 0  
    accuracy_score = ((TP+TN)/(TP+FP+TN+FN))*100
    recall_score = (TP/(TP+FN))*100
    
    cm.append(cf_matrix)
    acc.append(accuracy_score)
    recall.append(recall_score)
    presisi.append(precision_score)
    
    print(f"Fold {i+1}:")
    print(f"  Train :", len(X_train_fold))
    print(f"  Test :", len(X_test_fold))
    print(f"  Confusion Matrix")
    print(f"",cf_matrix)
    print(f"  Akurasi =", accuracy_score)
    print(f"  Presisi =", precision_score)
    print(f"  Recall =", recall_score)
    print("============================")
    
    # Simpan akurasi terbaik dan model yang sesuai
    if accuracy_score > best_acc:
        best_acc = accuracy_score
        best_model = classifier
        best_fold = i+1
```

```python
#Menampilkan confusion matrix rata-rata
import seaborn as sns
avg_cm = np.sum(cm, axis=0)
sns.heatmap(avg_cm, annot=True, cmap='Blues', fmt='g', xticklabels=['Negatif', 'Positif'], 
            yticklabels=['Negatif', 'Positif'])

#menambahkan label keterangan
plt.xlabel('Predicted Class')
plt.ylabel('Actual Class')
plt.title('Confusion Matrix')

print("Pengujian k-fold dengan seleksi fitur")
print("=================================================")
print("Rata-rata Akurasi :", np.mean(acc))
print("Rata-rata Presisi :", np.mean(presisi)) 
print("Rata-rata Recall :", np.mean(recall))
```

```python
#PENGUJIAN DATA TEST MENGGUNAKAN FOLD TERBAIK

cm_test=[]

# Hitung skor pada data test yang tersisa (20% dari seluruh data)
Test_X_final = matrixTransform(X_test)
y_pred = best_model.predict(Test_X_final)

#Confusion Matrix
#nilai 0 = negatif, nilai 1 = positif
cf_matrix = confusion_matrix(y_test, y_pred)
TN = cf_matrix[0][0] #prediksi salah, aktual salah
FN = cf_matrix[1][0] #prediksi benar, aktual salah
TP = cf_matrix[1][1] #prediksi benar, aktual benar
FP = cf_matrix[0][1] #prediksi salah, aktual benar
try :
    precision_score = (TP/(TP+FP))*100
except :
    precision_score = 0  
accuracy_score = ((TP+TN)/(TP+FP+TN+FN))*100
recall_score = (TP/(TP+FN))*100

cm_test.append(cf_matrix)
```
